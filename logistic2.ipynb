```python
import numpy as np
np.random.seed(0)
```

```python
N, D = 100, 2
```

```python
X = np.random.randn(N,D)
```

```python
X
```

```python
X[:50,:] = X[:50,:] - 2*np.ones((50,D))
X[50:,:] = X[50:,:] + 2*np.ones((50,D))
```

```python
Y = np.array([0]*50 + [1]*50)
```

```python
Y
```

```python
Y.shape
```

```python
ones = np.ones((N, 1))
Xb = np.concatenate((ones, X), axis=1)
w = np.random.randn(D + 1)

z = Xb.dot(w)
```

```python
z.shape
```

```python
def sigmoid(x):
    return 1.0/(1.0 + np.exp(-x))
```

```python
Yhat = sigmoid(z)
```

```python
Yhat
```

```python
def crossEntropy(Y, Yhat):
    E = 0
    for i in range(len(Y)):
        if Y[i] == 1:
            E -= np.log(Yhat[i])
        else:
            E -= np.log(1-Yhat[i])
    return E
```

```python
w = [0, 4, 4]
```

```python
z = Xb.dot(w)

Yhat = sigmoid(z)
```

```python
crossEntropy(Y, Yhat)
```
