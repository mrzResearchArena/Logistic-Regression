```python
import numpy as np
np.random.seed(0)
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def softplus(x):
    return np.log(1+np.exp(x))
```

```python
def relu(x):
    return x * (x > 0)
```

```python
v = np.arange(-10, 10)

plt.plot(v, sigmoid(v))
plt.grid()
plt.show()
```

```python
v = np.arange(-10, 10)

plt.plot(v, np.tanh(v))
plt.grid()
plt.show()
```

```python
v = np.arange(-10, 10)

plt.plot(v, softplus(v))
plt.grid()
plt.show()
```
